# Import libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Load a CSV file into a Pandas DataFrame:

```python
df = pd.read_csv("100 Sales Records.csv")
```

# Group by 'Region' and sum 'Total Revenue'

```python
regional_sales = df.groupby('Region')['Total Revenue'].sum()
```

# Plot the results (BAR)

```python
regional_sales.plot(kind='bar', figsize=(8, 6))
plt.title('Total Sales Amount by Region')
plt.xlabel('Region')
plt.ylabel('Total Sales Amount')
plt.xticks(rotation=45) # Rotate x-axis labels if they overlap
plt.tight_layout() # Adjust layout to prevent labels from being cut off
plt.show()
```

# Group by 'Country' and sum 'Total Profit'

```python
country_profit = df.groupby('Country')['Total Profit'].sum()
```

# Plot the results (Line)

```python
country_profit.plot(kind='line', figsize=(8, 6))
plt.title('Total Profit by Country')
plt.xlabel('Country')
plt.ylabel('Total Profit')
plt.xticks(rotation=45) # Rotate x-axis labels if they overlap
plt.tight_layout() # Adjust layout to prevent labels from being cut off
plt.show()
```

# Group by 'Sales Channel' and sum 'Units Sold'

```python
sales_channel_units = df.groupby('Sales Channel')['Units Sold'].sum()
```

# Plot the results (Pie)

```python
sales_channel_units.plot(kind='pie', figsize=(8, 6))
plt.title('Total Sales by Sales Channel')
plt.xlabel('Sales Channel')
plt.ylabel('Unit Sold')
plt.xticks(rotation=45) # Rotate x-axis labels if they overlap
plt.tight_layout() # Adjust layout to prevent labels from being cut off
plt.show()
```
